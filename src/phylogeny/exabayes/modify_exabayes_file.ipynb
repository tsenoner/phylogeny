```python
from pathlib import Path

# --- remove `{0}` part in each line starting with `\ttree`---`
path_dir = Path("../../../data/3FTx")
for exabayes_file in path_dir.glob("ExaBayes_topologies.run-*"):
    run_nr, identifier = [suffix[1:] for suffix in exabayes_file.suffixes]
    out_file = exabayes_file.with_name(f"{identifier}_{run_nr}.nexus")
    print(out_file)

    with open(exabayes_file) as handle_in, open(out_file, "w") as handle_out:
        for line in handle_in:
            if line.startswith("\ttree"):
                line = line.replace("{0}", "", 1)
            handle_out.write(line)
```

```python
import re

# --- extract all taxon names from the tree ---
nr_gen = 0
with (
    open(path_dir / "3FTx_resubmission_DSD_v2_run-0.nexus") as file_in,
    open(path_dir / "names.txt", "w") as file_out,
):
    taxons = False
    for line in file_in:
        if line.startswith("\ttranslate"):
            taxons = True
            continue
        if taxons:
            identifier = (
                re.match(r"\t\d+\t(.*)[,;]\n", line)
                .group(1)
                .split("|")[1]
                .replace("_", "")
                .replace(".", "")
                .replace("-", "")
            )
            file_out.write(f"{identifier}\n")
        if taxons and line.endswith(";\n"):
            taxons = False
        if line.startswith("\ttree"):
            nr_gen += 1

print(nr_gen)
```
